## Imports and Load Data

```python
#Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import json
```

```python
#Load data from 2000 file
df_2000 = pd.read_csv('Data/final_tmdb_data_2000.csv.gz')
df_2000.head()
```

```python
#Load data from 2001 file
df_2001 = pd.read_csv('Data/final_tmdb_data_2001.csv.gz')
df_2001.head()
```

## Combine both datasets into single dataframe and CSV file

```python
#Concatenate into one dataframe
df1 = pd.concat((df_2000, df_2001))
df1.info()
```

```python
FOLDER = "Data/"
df1.to_csv(f"{FOLDER}tmdb_results_combined.csv.gz")
```

## EDA

### 1. How many movies had at least some valid financial information (values > 0 for budget OR revenue)?


```python
budget_df = df1[(df1['budget'] > 0) | (df1['revenue'] > 0)]
print(f'The number of movies with budget or revenue greater than 0 is {len(budget_df)}.')
```

**Note: Please exclude any movies with 0's for budget AND revenue from the remaining visualizations.**

### 2. How many movies are there in each of the certification categories (G/PG/PG-13/R)?

```python
sns.countplot(data=budget_df, x='certification')
# Set the title and axis labels
plt.title('Movies by Certification Category')
plt.xlabel('Certification type')
plt.ylabel('Number of Movies')

# Show the plot
plt.show()
```

### 3. What is the average revenue per certification category?

```python
sns.barplot(data=budget_df, x='certification', y='revenue');

# Set the title and axis labels
plt.title('Average Revenue per Certification Type')
plt.xlabel('Certification Type')
plt.ylabel('Average Revenue')
```

### 4. What is the average budget per certification category?

```python
sns.barplot(data=budget_df, x='certification', y='budget');

# Set the title and axis labels
plt.title('Average Budget per Certification Type')
plt.xlabel('Certification Type')
plt.ylabel('Average Budget')
```
